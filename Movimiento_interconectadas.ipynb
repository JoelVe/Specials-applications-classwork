<h1>Movimiento relativo</h1>
En muchas ocasiones, a conveniencia de un ejercicio se vuelve prctico realizar el anlisis con respecto a un sistema de referencia auxiliar, y no a uno fijo. Existen diferentes problemas que se simplifican, definiendo ejes acordes al movimiento. Esto se denomina, movimiento relativo.

Para definir este sistema de coordenadas auxiliar, se debe usar como referencia un sistema de coordenadas fijo.

A partir de esta ayuda, se pueden encontrar consideraciones importantes, al tener un sistema con 2 o ms partculas, ya que se generan una serie de relaciones esenciales para el clculo.

Por ejemplo, si se tiene dos partculas en el espacio con movimientos diferentes, se puede definir un eje fijo a tierra y otro en una de las partculas, para de este modo encontrar las siguientes relaciones.



$${r_{PA}}={r_{BA}}+{r_{PB}}$$
Por definicin, es posible derivar la expresin resultante para de esta manera encontrar las relaciones de la velocidad y la aceleracin:

$$\dot{r_{PA}}=\dot{r_{BA}}+\dot{r_{PB}}$$
Equivalente a: ${v_{PA}}={v_{BA}}+{v_{PB}}$

Y la segunda derivada:

$$\ddot{r_{PA}}=\ddot{r_{BA}}+\ddot{r_{PB}}$$
Equivalente a: ${a_{PA}}={a_{BA}}+{a_{PB}}$

Dichas ecuaciones relacionan valores como la posicin, velocidad y la aceleracin entre una partcula y otra se vuelven muy importantes para resolver los diferentes problemas.

<h1>Ejercicio de aplicacin.</h1>
El sistema que se muestra comienza desde el reposo y cada componente se mueve con una aceleracin constante. Si la aceleracin relativa del bloque $C$ con respecto al collar es $60$  $mm/s$ hacia arriba y la aceleracin relativa del bloque $D$ con respecto al bloque A es $110$ $mm/s^2$ hacia abajo, determine:



(a) la velocidad del bloque C despus $3s$.


(b) El cambio de posicin del bloque $D$ despus de $5s$.

(c) Grafique la posicin en funcin del tiempo del bloque $D$ y comente el resultado.




 




Se coloca el punto de referencia $O$ en un punto fijo. Luego se colocan las posiciones de cada bloque con respecto al punto de referencia $O$, considerando que el eje positivo $y$ se encuentra hacia abajo. Se puede observar la existencia de dos cuerdas: la cuerda que pasa por las poleas de los bloques $ABC$ y la cuerda que pasa por las poleas de los bloques $ADB$.</n>




Analizando la cuerda $ABD$ se tiene:</n>

$$2y_A+2y_B+y_C+K_1=L_1$$

Donde $L_1$ es la longitud de la cuerda que se mantiene constante y $K_1$ es una constante. Derivando dos veces con respecto al tiempo se obtiene:

$${dL_1}{dt}=2\dot{y}_A+2\dot{y}_B+\dot{y}_C=0$$

$$\label{2} \frac{d^{2} L_1}{dt^2}=2\ddot{y}_A+2\ddot{y}_B+\ddot{y}_C=0 \quad (1)$$


Adems, se sabe que:

$$\label{1}\ddot{y}_{C/B}=-60\frac{mm}{s^2}=\ddot{y}_C-\ddot{y}_B \quad(2)  $$


$$\ddot{y}_{D/A}=110\frac{mm}{s^2}=\ddot{y}_D-\ddot{y}_A \quad(3)$$

Luego, analizando la cuerda $ABD$ se tiene:

$$y_D-y_B+y_D-y_A+K_2=L_2$$

Donde $L_2$ es la longitud de la cuerda y $K_2$ una constante.Derivando dos veces con respecto al tiempo se obtiene:

$$\frac{d^2L_2}{dt^2}=2\ddot{y}_D-\ddot{y}_B-\ddot{y}_A=0\quad(4)$$

Luego se obtiene el sistema de ecuaciones:

$\left\{\begin{matrix}
2\ddot{y}_A+2\ddot{y}_B+\ddot{y}_C=0\\ 
\ddot{y}_C-\ddot{y}_B=-60\\ 
\ddot{y}_D-\ddot{y}_A=110\\ 
2\ddot{y}_D-\ddot{y}_B-\ddot{y}_A=0
\end{matrix}\right.$


De donde se obtiene:
$$\ddot{y}_A=-120\frac{mm}{s^2}$$

$$\ddot{y}_B=100\frac{mm}{s^2}$$

$$\ddot{y}_C=40\frac{mm}{s^2}$$

$$\ddot{y}_D=-10\frac{mm}{s^2}$$

Como las aceleraciones de todo el sistema son constantes, se trata de un movimiento rectilneo uniforme. Entonces, como el sistema se mueve desde el reposo, se tiene:


$$\dot{y}_{C_{t=3}}=\dot{y}_{C_{t=0}}+\ddot{y}_Ct_0=0+40\times(3)$$

$$\dot{y}_{C_{t=3}}=120\frac{mm}{s}$$

Por tanto, $C$ est movindose hacia abajo. Por otra parte:

$$\Delta y_{D_{t=5}}=\dot{y}_{D_{t=0}}t+\frac{1}{2}\ddot{y}_Dt^2=0+\frac{1}{2}(-10)(5^2)$$

$$\Delta y_{D_{t=5}}=-125mm$$

Por tanto, el desplazamiento de C es hacia arriba. Finalmente, se obtiene el desplazamiento de D en funcin del tiempo:

$$\Delta y_{D}=-5t^2$$

Donde $\Delta y_{D}$ est en $[mm]$ y $[t]$ en segundos.

```python
# **Grfico** 
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go

t = np.linspace(0,10,100)
s = -5*t**2
fig=px.line(x=t,y=s,labels={'x':'Tiempo [s]','y':'Desplazamiento [mm]'},title="Desplazamiento vs tiempo del bloque D")
fig.show()
```
